### Amount of Playground Equipment

My data sets primary goal is to analyze the recorded amount of playground equipment in each of the neighborhoods and determine the one with the most equiped playground. My thought behind looking at this data is to prove what the best and most memorable neighborhood to raise a kid would be, because it would provide the kid with the most equipment to play with and give them good memories. I started by using a dataset that collected data from parks in all Pittsburgh neighborhoods and has been consistently updated.


```python
#Importing Pandas and other tools
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

#Importing the original csv
play_og = pd.read_csv("playground_eqp.csv")
play_og.head(10)
```

There were too many unnecessary columns that I wouldn't be using. So to make it more readable I removed some of them.

```python
#Removing some of the unnecessary columns
play = play_og[["id","name", "neighborhood"]]
play.head(10)
```

I decided to use a dictionary to add up the total number of equipment in each of the neighborhoods. It was the easiest way to keep the amount of equipment linked to the neighborhood name because of the keys and values. 

```python
#Creating the Dictionary
nei_sum = {}
#Iterating through the dataframe and summing the equipment up
for index, row in play.iterrows():
    if(row["neighborhood"] in nei_sum):
        nei_sum[row["neighborhood"]] = (nei_sum.get(row["neighborhood"])) + 1
    else:
        nei_sum[row["neighborhood"]] = 1
#Displaying the dictionary with the total sum for each neighborhood
print(nei_sum)
    
```

After getting everything into the dictionary I wanted to sort the data from highest to lowest so that the data was more easily readible. 

```python
nei_sort = {}

#For loop to iterate through the keys and values
for k in sorted(nei_sum, key=nei_sum.get, reverse=True):
    nei_sort[k] = nei_sum.get(k)

#Displaying the sorted dictionary
print(nei_sort)
```

Using the sorted dictionary I then converted it back into a dataframe to be able to make it into a bar graph eventually. 

```python
#Moving it into the data frame and setting the column names
sortedF = pd.DataFrame(nei_sort.items())
sortedF.columns = ["Neighborhood", "Sum of Equipment"]

sortedF.head(10)
```

Displaying the top 15 neighborhoods from ascending order on a bar graph

```python
#Setting how big the graph can be
plt.rcParams["figure.figsize"] = (20,10)

#Plotting the graph with neighborhoods being on x and number of equipment being on y
sortedF.head(15).plot.bar(x="Neighborhood", y="Sum of Equipment", rot=70, title="Sum of Playground Equipment per Neighborhood")
plt.show(block=True)
```

Displaying the worst 10 neighborhoods

```python
#Plotting the graph with the worst neighborhoods being on x and number of equipment being on y
sortedF.tail(10).plot.bar(x="Neighborhood", y="Sum of Equipment", rot=70, title="Equipment of Playground Equipment per Neighborhood")
plt.show(block=True)
```

```python
enrollmentData= pd.read_csv("enrollmentData.csv")
enrollmentData.head()
```

**Since there 40 columns, it would take a long time to type in and delete them so I created a new DataFrame called enrollmentData1.**



```python
##printing the columns that will be used in the obervation
enrollmentData1=enrollmentData[["Neighborhood","Estimate; Enrolled in school:","Estimate; Enrolled in school: - Enrolled in college, undergraduate years","Estimate; Not enrolled in school"]].copy()
enrollmentData1.head(10)
```

**I used the columns pulled from the dataFrame to create new columns based off their stats, and then created a ranking system.**

```python
##Adding Estimated Enrolled, to Estimated not enrolled for student total. Then Finding the enrollment Percentage. 
##Adding percentage of students who go on to get a college degree
##Adding a ranking system to the percentages, Low being a bad score, Medium being a average score, and High being a above average score.
def my_function(x):
    if x>.5:
        return("high")
    if x>.3:
        return("medium")
    return("low")

result=enrollmentData1.iloc[:,[1,3]].sum(axis=1)

enrollmentData1["Enrollment Percentage Overall"] = ((enrollmentData1["Estimate; Enrolled in school:"])/result)
enrollmentData1["Enrollment Percentage undergradute degree"] = (enrollmentData1["Estimate; Enrolled in school: - Enrolled in college, undergraduate years"]/result)
enrollmentData1["Enrollment Percentage Overall, high, medium, low"]=enrollmentData1["Enrollment Percentage Overall"].map(my_function)
enrollmentData1["Enrollment Percentage undergraduate status, high, medium, low"]=enrollmentData1["Enrollment Percentage undergradute degree"].map(my_function)


enrollmentData1.head(10)
##Printing Out the First 10
```
